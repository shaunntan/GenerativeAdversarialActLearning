```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import usps.usps as usps
```

```python

(x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()
y_train = np.squeeze(y_train)
y_test = np.squeeze(y_test)
x_train_1 = (x_train[(y_train == 1),])
y_train_1 = y_train[(y_train == 1)]
x_train_7 = (x_train[(y_train == 7),])
y_train_7 = y_train[(y_train == 7)]

for i in range(4 * 4):
    # define subplot
    plt.subplot(4, 4, 1 + i)
    # turn off axis
    plt.axis('off')
    # plot raw pixel data
    if (i <= 7):
        plt.imshow(x_train_1[np.random.randint(0,5000)].astype('uint8'))
    else:
        plt.imshow(x_train_7[np.random.randint(0,5000)].astype('uint8'))
```

```python
(x_test, y_test), (_, _) = usps.load_data()

y_test = y_test.astype('int')
x_test_5 = x_test[(y_test == 5),]
x_test_7 = x_test[(y_test == 7),]

a = x_test_5[0]

x_test_5_b = np.expand_dims(x_test_5, axis=3)
x_test_5_b = (x_test_5_b + 1) / 2
x_test_5_b = tf.image.resize(x_test_5_b, [28,28])

b = x_test_5_b[0]

x_test_5_c = np.expand_dims(x_test_5, axis=3)
x_test_5_c = (x_test_5_c + 1) / 2
x_test_5_c = tf.image.pad_to_bounding_box(x_test_5_c,4,4,24,24)
x_test_5_c = tf.image.resize(x_test_5_c, [28,28])
x_test_5_c = (x_test_5_c * 2) - 1

c = x_test_5_c[0]

(x_train, y_train), (_, _) = tf.keras.datasets.mnist.load_data()
x_train_mnist5 = x_train[(y_train == 5),]

d = x_train_mnist5[7]
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1, 7, 1)
plt.axis('off')
plt.imshow(a, cmap = 'gray')
plt.title('USPS:\nOriginal 16x16')
plt.subplot(1, 7, 3)
plt.axis('off')
plt.imshow(b, cmap = 'gray')
plt.title('USPS:\nUpsampled 28x28\nwithout Padding')
plt.subplot(1, 7, 5)
plt.axis('off')
plt.imshow(c, cmap = 'gray')
plt.title('USPS:\nUpsampled 28x28\nwith Padding')
plt.subplot(1, 7, 7)
plt.axis('off')
plt.imshow(d, cmap = 'gray')
plt.title('MNIST')
plt.gcf().set_dpi(300)
```
