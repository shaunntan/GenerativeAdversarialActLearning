```python
import dill as pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
def tidyresults_notfs(r):
    means = np.mean(r[1], axis = 0)
    sd = np.std(r[1], axis=0)
    return means, sd
```

# Train and test on mnist

```python
with open('./result_gaal_mnistmnist.pkl', 'rb') as file:
    result_gaal = pickle.load(file)
    
with open('./result_activesvm_mnistmnist.pkl', 'rb') as file:
    result_activesvm = pickle.load(file)
    
with open('./result_random_mnistmnist.pkl', 'rb') as file:
    result_random = pickle.load(file)

with open('./result_simplegan_mnistmnist.pkl', 'rb') as file:
    result_simplegan = pickle.load(file)
    
with open('./result_fs_mnistmnist.pkl', 'rb') as file:
    result_fs = pickle.load(file)
    
    
means_gaal, sd_gaal = tidyresults_notfs(result_gaal)  
means_asvm, sd_asvm = tidyresults_notfs(result_activesvm)
means_random, sd_random = tidyresults_notfs(result_random)
means_simplegan, sd_simplegan = tidyresults_notfs(result_simplegan)  

plt.title('Train and test on MNIST 5 & 7')
plt.axhline(y=result_fs[1], color='r', linestyle='--', label="Fully Supervised")
plt.errorbar(result_activesvm[2][0], y=means_asvm, yerr=sd_asvm, color='purple', ls='--', alpha=0.8, capsize=2, label="Active SVM")
plt.errorbar(result_gaal[2][0], y=means_gaal, yerr=sd_gaal, color='k', alpha=0.8, capsize=2, label="GAAL")
plt.errorbar(result_simplegan[2][0], y=means_simplegan, yerr=sd_simplegan, color='g', alpha=0.8, capsize=2, label="Simple GAN")
plt.errorbar(result_random[2][0], y=means_random, yerr=sd_random, color='b', ls = '--', alpha=0.8, capsize=2, label="Random Sampling")
plt.legend(loc='lower right')
plt.ylim([0.7,1.0])
plt.ylabel('Classification Accuracy')
plt.xlabel('Number of labelled samples')
plt.grid()
plt.gcf().set_dpi(300)
plt.tight_layout()
plt.show()
```

# Compare Quality of Training with Generator from DCGAN vs WGAN

```python
with open('./result_gaal_cifar10.pkl', 'rb') as file:
    result_gaal_cifar10 = pickle.load(file)

with open('./result_gaal_cifar10_wgan.pkl', 'rb') as file:
    result_gaal_cifar10_wgan = pickle.load(file)
    
with open('./result_gaal_cifar10_3000.pkl', 'rb') as file:
    result_gaal_cifar10_3000 = pickle.load(file)

with open('./result_gaal_cifar10_wgan_3000.pkl', 'rb') as file:
    result_gaal_cifar10_wgan_3000 = pickle.load(file)
    
with open('./result_gaal_cifar10_1500_d.pkl', 'rb') as file:
    result_gaal_cifar10_1500_d = pickle.load(file)

with open('./result_gaal_cifar10_3000_d.pkl', 'rb') as file:
    result_gaal_cifar10_3000_d = pickle.load(file)
    
with open('./result_gaal_cifar10_wgan_1500_d.pkl', 'rb') as file:
    result_gaal_cifar10_wgan_1500_d = pickle.load(file)

with open('./result_gaal_cifar10_wgan_3000_d.pkl', 'rb') as file:
    result_gaal_cifar10_wgan_3000_d = pickle.load(file)

with open('./result_fs_cifar10.pkl', 'rb') as file:
    result_fs = pickle.load(file)

means_gaal_1500, sd_gaal_1500 = tidyresults_notfs(result_gaal_cifar10)
means_gaal_3000, sd_gaal_3000 = tidyresults_notfs(result_gaal_cifar10_3000)
means_gaal_1500_wgan, sd_gaal_1500_wgan = tidyresults_notfs(result_gaal_cifar10_wgan)
means_gaal_3000_wgan, sd_gaal_3000_wgan = tidyresults_notfs(result_gaal_cifar10_wgan_3000)

means_gaal_1500_d, sd_gaal_1500_d = tidyresults_notfs(result_gaal_cifar10_1500_d)
means_gaal_3000_d, sd_gaal_3000_d = tidyresults_notfs(result_gaal_cifar10_3000_d)
means_gaal_1500_wgan_d, sd_gaal_1500_wgan_d = tidyresults_notfs(result_gaal_cifar10_wgan_1500_d)
means_gaal_3000_wgan_d, sd_gaal_3000_wgan_d = tidyresults_notfs(result_gaal_cifar10_wgan_3000_d)

plt.title('GAAL: Train and test on CIFAR-10\nGenerator from DCGAN vs WGAN')
plt.axhline(y=result_fs[1], color='purple', linestyle='--', label="Fully Supervised")
plt.errorbar(result_gaal_cifar10[2][0], y=means_gaal_1500, yerr=sd_gaal_1500, color='r', alpha=0.8, capsize=2, label="DCGAN 1500 Epochs")
plt.errorbar(result_gaal_cifar10_wgan[2][0], y=means_gaal_1500_wgan, yerr=sd_gaal_1500_wgan, color='b', alpha=0.8, capsize=2, label="WGAN 1500 Epochs")
# plt.errorbar(result_gaal_cifar10_3000[2][0], y=means_gaal_3000, yerr=sd_gaal_3000, color='g', alpha=0.8, capsize=2, label="DCGAN 3000 Epochs")
# plt.errorbar(result_gaal_cifar10_wgan_3000[2][0], y=means_gaal_3000_wgan, yerr=sd_gaal_3000_wgan, color='black', alpha=0.8, capsize=2, label="WGAN 3000 Epochs")
plt.legend(title='Type of GAN training (Number of Generator Epochs)', loc='lower right')
# plt.ylim([0.55,0.85])
plt.ylabel('Classification Accuracy')
plt.xlabel('Number of labelled samples')
plt.grid()
plt.gcf().set_dpi(300)
plt.tight_layout()
plt.show()
```

# Diversity

```python
with open(f'./result_gaal_cifar10.pkl', 'rb') as file:
    result = pickle.load(file)

means, sd = tidyresults_notfs(result)

plt.errorbar(result[2][0], y=means, yerr=sd, color = 'black', alpha=0.8, capsize=2, label="No Penalty")

with open('./result_fs_cifar10.pkl', 'rb') as file:
    result_fs = pickle.load(file)
    
plt.axhline(y=result_fs[1], color='purple', linestyle='--', label="Fully Supervised")

ls = [0.0001, 0.001, 0.01, 0.1, 1]

for l in ls:
    with open(f'./result_gaal_cifar10_d_{l}.pkl', 'rb') as file:
        result = pickle.load(file)
        
    means, sd = tidyresults_notfs(result)

    plt.errorbar(result[2][0], y=means, yerr=sd, alpha=0.8, capsize=2, label=f"{l}")

plt.title('GAAL with Diversity (Average Distance) Penalty (CIFAR-10)')
plt.legend(title='Penalty', bbox_to_anchor=(1.04,0), loc="lower left", borderaxespad=0)
# plt.ylim([0.5,0.8])
plt.ylabel('Classification Accuracy')
plt.xlabel('Number of labelled samples')
plt.grid()
# plt.figure(figsize=(160, 120), dpi=300)
plt.gcf().set_dpi(300)
plt.gcf().set_size_inches(7,3)
plt.tight_layout()
plt.show()
```
